```python
import os
import glob
import random
import numpy as np
import tensorflow as tf
import cv2
import matplotlib.pyplot as plt

# ============================================================
# CONFIGURATION
# ============================================================
IMAGE_DIRECTORY = 'data/images/train/'
MASK_DIRECTORY  = 'data/images/masks/'

IMG_SIZE = 256
BATCH    = 8
EPOCHS   = 50
LR       = 1e-3

# Train / Val / Test split ratios
TRAIN_RATIO = 0.80
VAL_RATIO   = 0.10
# TEST_RATIO  = 0.10 (remainder)

RANDOM_SEED = 42

# ============================================================
# IO HELPERS
# ============================================================
def read_gray(path: str):
    img = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if img is None:
        raise FileNotFoundError(f"Cannot read image: {path}")
    img = cv2.resize(img, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_AREA)
    img = img.astype(np.float32) / 255.0
    return img[..., None]   # (H, W, 1)

def read_mask(path: str):
    if (not path) or (not os.path.exists(path)):
        return np.zeros((IMG_SIZE, IMG_SIZE, 1), dtype=np.float32)
    m = cv2.imread(path, cv2.IMREAD_GRAYSCALE)
    if m is None:
        return np.zeros((IMG_SIZE, IMG_SIZE, 1), dtype=np.float32)
    m = cv2.resize(m, (IMG_SIZE, IMG_SIZE), interpolation=cv2.INTER_NEAREST)
    m = (m > 0).astype(np.float32)
    return m[..., None]

# ============================================================
# DATA LOADING  вҖ”  shuffled before split to avoid leakage
# ============================================================
def load_local_data(seed=RANDOM_SEED):
    img_paths  = sorted(glob.glob(os.path.join(IMAGE_DIRECTORY, "*.png")))

    valid_imgs, valid_masks = [], []
    for p in img_paths:
        fname  = os.path.basename(p)
        m_path = os.path.join(MASK_DIRECTORY, fname)
        if os.path.exists(m_path):
            valid_imgs.append(p)
            valid_masks.append(m_path)

    # --- FIX: shuffle BEFORE splitting to prevent data leakage ---
    combined = list(zip(valid_imgs, valid_masks))
    random.seed(seed)
    random.shuffle(combined)
    valid_imgs, valid_masks = zip(*combined) if combined else ([], [])

    return list(valid_imgs), list(valid_masks)


def split_data(img_paths, mask_paths):
    n          = len(img_paths)
    train_end  = int(TRAIN_RATIO * n)
    val_end    = int((TRAIN_RATIO + VAL_RATIO) * n)

    train_imgs,  train_masks  = img_paths[:train_end],  mask_paths[:train_end]
    val_imgs,    val_masks    = img_paths[train_end:val_end], mask_paths[train_end:val_end]
    test_imgs,   test_masks   = img_paths[val_end:],    mask_paths[val_end:]

    print(f"  Train : {len(train_imgs)} samples")
    print(f"  Val   : {len(val_imgs)}   samples")
    print(f"  Test  : {len(test_imgs)}  samples")

    return (train_imgs, train_masks), (val_imgs, val_masks), (test_imgs, test_masks)

# ============================================================
# AUGMENTATION  вҖ”  applied only during training
# ============================================================
def augment(img, mask):
    # Horizontal flip
    if tf.random.uniform([]) < 0.5:
        img  = tf.image.flip_left_right(img)
        mask = tf.image.flip_left_right(mask)
    # Vertical flip
    if tf.random.uniform([]) < 0.5:
        img  = tf.image.flip_up_down(img)
        mask = tf.image.flip_up_down(mask)
    # Random 90-degree rotation
    k    = tf.random.uniform([], 0, 4, dtype=tf.int32)
    img  = tf.image.rot90(img,  k)
    mask = tf.image.rot90(mask, k)
    # Photometric jitter (image only)
    img  = tf.image.random_brightness(img, 0.05)
    img  = tf.image.random_contrast(img, 0.9, 1.1)
    img  = tf.clip_by_value(img, 0.0, 1.0)
    return img, mask

# ============================================================
# tf.data PIPELINE
# ============================================================
def tf_loader(img_path, mask_path, training=True):
    def _py(img_p, mask_p):
        img = read_gray(img_p.decode())
        msk = read_mask(mask_p.decode())
        return img, msk

    img, msk = tf.numpy_function(_py, [img_path, mask_path], [tf.float32, tf.float32])
    img.set_shape([IMG_SIZE, IMG_SIZE, 1])
    msk.set_shape([IMG_SIZE, IMG_SIZE, 1])

    if training:
        img, msk = augment(img, msk)
    return img, msk


def make_ds(img_paths, mask_paths, training=True):
    ds = tf.data.Dataset.from_tensor_slices((img_paths, mask_paths))
    if training:
        ds = ds.shuffle(512, seed=RANDOM_SEED)
    ds = ds.map(
        lambda a, b: tf_loader(a, b, training),
        num_parallel_calls=tf.data.AUTOTUNE
    )
    ds = ds.batch(BATCH).prefetch(tf.data.AUTOTUNE)
    return ds

# ============================================================
# LIGHTWEIGHT U-NET  вҖ”  base=32 for better capacity
# ============================================================
def conv_block(x, f):
    x = tf.keras.layers.Conv2D(f, 3, padding="same")(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    x = tf.keras.layers.Conv2D(f, 3, padding="same")(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.ReLU()(x)
    return x


def build_light_unet(base=32):
    inp = tf.keras.Input((IMG_SIZE, IMG_SIZE, 1))

    # --- Encoder ---
    c1 = conv_block(inp, base);       p1 = tf.keras.layers.MaxPool2D()(c1)
    c2 = conv_block(p1,  base * 2);  p2 = tf.keras.layers.MaxPool2D()(c2)
    c3 = conv_block(p2,  base * 4);  p3 = tf.keras.layers.MaxPool2D()(c3)
    c4 = conv_block(p3,  base * 8);  p4 = tf.keras.layers.MaxPool2D()(c4)

    # --- Bottleneck ---
    bn = conv_block(p4, base * 16)

    # --- Decoder ---
    u4 = tf.keras.layers.Conv2DTranspose(base * 8,  2, strides=2, padding="same")(bn)
    u4 = tf.keras.layers.Concatenate()([u4, c4]);  c5 = conv_block(u4, base * 8)

    u3 = tf.keras.layers.Conv2DTranspose(base * 4,  2, strides=2, padding="same")(c5)
    u3 = tf.keras.layers.Concatenate()([u3, c3]);  c6 = conv_block(u3, base * 4)

    u2 = tf.keras.layers.Conv2DTranspose(base * 2,  2, strides=2, padding="same")(c6)
    u2 = tf.keras.layers.Concatenate()([u2, c2]);  c7 = conv_block(u2, base * 2)

    u1 = tf.keras.layers.Conv2DTranspose(base,      2, strides=2, padding="same")(c7)
    u1 = tf.keras.layers.Concatenate()([u1, c1]);  c8 = conv_block(u1, base)

    out = tf.keras.layers.Conv2D(1, 1, activation="sigmoid")(c8)
    return tf.keras.Model(inp, out, name="LightUNet256")

# ============================================================
# METRICS & LOSS
# ============================================================
def dice_coef(y_true, y_pred, eps=1e-6):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.clip_by_value(tf.cast(y_pred, tf.float32), 0.0, 1.0)
    inter  = tf.reduce_sum(y_true * y_pred, axis=[1, 2, 3])
    denom  = tf.reduce_sum(y_true + y_pred, axis=[1, 2, 3])
    return tf.reduce_mean((2.0 * inter + eps) / (denom + eps))


def combo_loss(y_true, y_pred):
    bce    = tf.keras.losses.BinaryCrossentropy()(y_true, y_pred)
    dice_l = 1.0 - dice_coef(y_true, y_pred)
    # FIX: weight Dice more heavily to handle class imbalance (small nerve regions)
    return 0.3 * bce + 0.7 * dice_l


# ============================================================
# PLOTTING
# ============================================================
def plot_training_history(history, save_path="training_charts.png"):
    epochs_ran = range(1, len(history.history["loss"]) + 1)

    fig, axes = plt.subplots(1, 3, figsize=(18, 5))
    fig.suptitle("Training History", fontsize=16, fontweight="bold")

    # Chart 1: Loss
    axes[0].plot(epochs_ran, history.history["loss"],     label="Train Loss", color="royalblue")
    axes[0].plot(epochs_ran, history.history["val_loss"], label="Val Loss",   color="tomato", linestyle="--")
    axes[0].set_title("Loss (Combo: BCE + Dice)")
    axes[0].set_xlabel("Epoch"); axes[0].set_ylabel("Loss")
    axes[0].legend(); axes[0].grid(True, alpha=0.3)

    # Chart 2: Dice Coefficient
    axes[1].plot(epochs_ran, history.history["dice_coef"],     label="Train Dice", color="royalblue")
    axes[1].plot(epochs_ran, history.history["val_dice_coef"], label="Val Dice",   color="tomato", linestyle="--")
    axes[1].set_title("Dice Coefficient")
    axes[1].set_xlabel("Epoch"); axes[1].set_ylabel("Dice"); axes[1].set_ylim([0, 1])
    axes[1].legend(); axes[1].grid(True, alpha=0.3)

    # Chart 3: IoU
    axes[2].plot(epochs_ran, history.history["iou"],     label="Train IoU", color="royalblue")
    axes[2].plot(epochs_ran, history.history["val_iou"], label="Val IoU",   color="tomato", linestyle="--")
    axes[2].set_title("Binary IoU")
    axes[2].set_xlabel("Epoch"); axes[2].set_ylabel("IoU"); axes[2].set_ylim([0, 1])
    axes[2].legend(); axes[2].grid(True, alpha=0.3)

    plt.tight_layout()
    plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()
    print(f"Training charts saved to: {save_path}")


def plot_predictions(model, test_ds, num_samples=4, save_path="prediction_samples.png"):
    """Show: input image | ground truth mask | predicted mask side by side."""
    images, masks = [], []
    for img_batch, mask_batch in test_ds.take(1):
        images = img_batch[:num_samples].numpy()
        masks  = mask_batch[:num_samples].numpy()
        break

    preds = model.predict(tf.constant(images), verbose=0)

    fig, axes = plt.subplots(num_samples, 3, figsize=(10, num_samples * 3))
    fig.suptitle("Prediction Samples", fontsize=13, fontweight="bold")

    col_titles = ["Input Image", "Ground Truth", "Prediction (>0.5)"]
    for col, title in enumerate(col_titles):
        axes[0][col].set_title(title, fontsize=11)

    for i in range(num_samples):
        axes[i][0].imshow(images[i, ..., 0], cmap="gray")
        axes[i][1].imshow(masks[i,  ..., 0], cmap="gray", vmin=0, vmax=1)
        axes[i][2].imshow((preds[i, ..., 0] > 0.5).astype(np.float32), cmap="gray", vmin=0, vmax=1)
        for ax in axes[i]:
            ax.axis("off")

    plt.tight_layout()
    plt.savefig(save_path, dpi=150, bbox_inches="tight")
    plt.show()
    print(f"Prediction samples saved to: {save_path}")

# ============================================================
# MAIN
# ============================================================
def main():
    # 1. Load & shuffle data
    all_imgs, all_masks = load_local_data()
    print(f"\nTotal matched samples: {len(all_imgs)}")
    if len(all_imgs) == 0:
        raise RuntimeError("No matching images/masks found. Check your folder paths!")

    # 2. Split into train / val / test
    (train_imgs, train_masks), \
    (val_imgs,   val_masks),   \
    (test_imgs,  test_masks)   = split_data(all_imgs, all_masks)

    # 3. Build tf.data datasets
    train_ds = make_ds(train_imgs, train_masks, training=True)
    val_ds   = make_ds(val_imgs,   val_masks,   training=False)
    test_ds  = make_ds(test_imgs,  test_masks,  training=False)

    # 4. Build model  вҖ”  FIX: base=32 for better capacity on fine nerve structures
    model = build_light_unet(base=32)
    model.summary()

    # 5. Compile with Dice + binary IoU metrics
    model.compile(
        optimizer=tf.keras.optimizers.Adam(LR),
        loss=combo_loss,
        metrics=[
            dice_coef,
            tf.keras.metrics.BinaryIoU(target_class_ids=[1], threshold=0.5, name="iou")
        ]
    )

    # 6. Callbacks
    callbacks = [
        tf.keras.callbacks.ModelCheckpoint(
            "light_unet_best.keras",
            monitor="val_dice_coef", mode="max",
            save_best_only=True, verbose=1
        ),
        # FIX: patience=10 is more sensible relative to 50 epochs
        tf.keras.callbacks.EarlyStopping(
            monitor="val_dice_coef", mode="max",
            patience=40, restore_best_weights=True, verbose=1
        ),
        # FIX: reduce LR when plateau detected
        tf.keras.callbacks.ReduceLROnPlateau(
            monitor="val_dice_coef", mode="max",
            factor=0.5, patience=5,
            min_lr=1e-6, verbose=1
        ),
        tf.keras.callbacks.CSVLogger("training_log.csv")
    ]

    # 7. Train
    print("\n--- Training ---")
    history = model.fit(
        train_ds,
        validation_data=val_ds,
        epochs=EPOCHS,
        callbacks=callbacks
    )

    # 8. Evaluate on held-out test set
    print("\n--- Test Set Evaluation ---")
    results = model.evaluate(test_ds, verbose=1)
    for name, val in zip(model.metrics_names, results):
        print(f"  {name}: {val:.4f}")

    # 9. Plot training curves
    print("\n--- Generating Charts ---")
    plot_training_history(history, save_path="training_charts.png")

    # 10. Visualize predictions on test samples
    plot_predictions(model, test_ds, num_samples=4, save_path="prediction_samples.png")

    print("\nBest model saved to: light_unet_best.keras")
    print("Training log saved to: training_log.csv")


if __name__ == "__main__":
    main()
```
